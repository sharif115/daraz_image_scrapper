```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Replace this path with your ChromeDriver path
service = Service("C:\WebDriver\chromedriver-win64\chromedriver-win64")
driver = webdriver.Chrome(service=service)

base_url = "https://www.daraz.com.bd/catalog/?q=headphone&page={}"
driver.get(base_url.format(1))
time.sleep(3)

# Find total pages
try:
    pages = driver.find_elements(By.CLASS_NAME, "ant-pagination-item")
    total_pages = int(pages[-1].text) if pages else 1
except Exception:
    total_pages = 1

print(f"Total Pages Found: {total_pages}")

products_all = []

# Loop through pages
for page in range(1, total_pages + 1):
    print(f"Scraping Page {page}...")
    driver.get(base_url.format(page))
    time.sleep(3)
    
    titles = driver.find_elements(By.CLASS_NAME, "title--wFj93")
    for title in titles:
        products_all.append(title.text)

driver.quit()

# Save to CSV
df = pd.DataFrame(products_all, columns=["Product Name"])
df.to_csv("daraz_products.csv", index=False)
print("Saved to daraz_products.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd

#Step 1: Correct ChromeDriver path
service = Service(r"C:\WebDriver\chromedriver-win64\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service)

#Step 2: Daraz URL
base_url = "https://www.daraz.com.bd/catalog/?q=headphone&page={}"
driver.get(base_url.format(1))
time.sleep(3)

#Step 3: Get total number of pages
try:
    pages = driver.find_elements(By.CLASS_NAME, "ant-pagination-item")
    total_pages = int(pages[-1].text) if pages else 1
except Exception:
    total_pages = 1

print(f"Total Pages Found: {total_pages}")

#Step 4: Scrape product titles from all pages
products_all = []

for page in range(1, total_pages + 1):
    print(f"Scraping Page {page}...")
    driver.get(base_url.format(page))
    time.sleep(3)
    
    titles = driver.find_elements(By.CLASS_NAME, "title--wFj93")
    for title in titles:
        products_all.append(title.text)

#Step 5: Quit browser
driver.quit()

# Step 6: Save to CSV
df = pd.DataFrame(products_all, columns=["Product Name"])
df.to_csv("daraz_products.csv", index=False)
print("Data saved to daraz_products.csv")
```
